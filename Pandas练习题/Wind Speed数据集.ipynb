# Wind Speed 数据集

### Step 1. 导入库

```python
import numpy as np
import pandas as pd
import datetime
import ssl
ssl._create_default_https_context = ssl._create_unverified_context # 全局取消证书验证
```

### Step 2. 导入数据集 [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data)

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
wind = pd.read_table(url, sep='\s+')
wind_copy = wind.copy()
```

### Step 3. 将前三列设置成datetime，修改其中错误，并作为索引。

#### 一种解法

```python
wind['Date'] = wind.Yr.astype(str) + '-' + wind.Mo.astype(str) + '-' + wind.Dy.astype(str) # 合并年月日成一列
wind['Date'] = pd.to_datetime(wind['Date'])   # 转换成datetime，观察到2061年，将其改为1961年

wind['Date'] = wind['Date'].astype(str).str.replace('20', '19') # 将Datetime转换成str，并将2061换成1961
wind = wind.set_index(pd.to_datetime(wind['Date']))
wind = wind.drop(['Yr', 'Mo', 'Dy', 'Date'], axis=1)
wind.head()
```

#### 另一种解法

```python
wind_copy['Date'] = wind_copy.Yr.astype(str) + '-' + wind_copy.Mo.astype(str) + '-' + wind_copy.Dy.astype(str) # 合并年月日成一列
wind_copy['Date'] = pd.to_datetime(wind_copy['Date'])  # 转换成datetime，观察到2061年，将其改为1961年,不光2061

def fix_year(x):
  year = x.year - 100 if x.year > 1989 else x.year  # 重命名一个新变量，而不是x.year
  return datetime.date(year, x.month, x.day)

# apply the function fix_century on the column and replace the values to the right ones
wind_copy['Date'] = wind_copy['Date'].apply(fix_year) # index没有.apply方法

# data.info()
wind_copy.head()
```

### Step 6. 创建一个DataFrame，叫loc_stats，统计不同地方风速的最小值、最大值、均值、方差

```python
loc_stats = pd.DataFrame()
loc_stats['min'] = wind.min(axis=0)
loc_stats['max'] = wind.max(axis=0)
loc_stats['mean'] = wind.mean(axis=0)
loc_stats['std'] = wind.std(axis=0)
loc_stats
```

### Step 7. 创建一个DataFrame，叫day_stats，统计每天风速的最小值、最大值、均值、方差

```python
day_stats = pd.DataFrame()
day_stats['min'] = wind.min(axis=1)
day_stats['max'] = wind.max(axis=1)
day_stats['mean'] = wind.mean(axis=1)
day_stats['std'] = wind.std(axis=1)
day_stats.head()
```
