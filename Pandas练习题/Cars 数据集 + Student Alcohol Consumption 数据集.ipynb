# Student Alcohol Consumption 数据集

### Step 1. 导入数据，并命名为df：[address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv).

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_table(url, sep=',')
```

### Step 2. 选择 'school' 列到 'guardian' 列为新的df

```python
df = df.loc[:, 'school': 'guardian']
df.head()
```

### Step 3. 大写 Mjob 列和Fjob列

```python
df['Mjob'] = df['Mjob'].apply(lambda x: x.capitalize())
df['Fjob'] = df['Fjob'].apply(lambda x: x.capitalize())
df.head()
```

### Step 4. 新建一列根据年龄判断能否喝酒（17岁为界限）

```python
df['legal_drinker'] = np.where(df['age']>17, True, False)
```

### Step 5. 将数据集中数字乘以10

```python
def f(x):
    if type(x) == int or  type(x) == float:
        return x*10
    else:
        return x
```

```python
df.applymap(f)
```
