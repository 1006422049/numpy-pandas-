# Fictitious Names 数据集

### Step 1. 根据raw_data创建DataFrame，命名为data1, data2, data3

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
data1 = pd.DataFrame(raw_data_1, columns=['subject_id', 'first_name', 'last_name'])
data2 = pd.DataFrame(raw_data_2, columns=['subject_id', 'first_name', 'last_name'])
data3 = pd.DataFrame(raw_data_3, columns=['subject_id', 'test_id'])
```

### Step 2. 按列名合并data1，data2，并命名为 all_data

```python
all_data = pd.concat([data1, data2], axis=0)
```

### Step 3. 沿着行合并data1，data2，并命名为all_data_col

```python
all_data_col = pd.concat([data1, data2], axis=1)
```

### Step 4. 根据subject_id合并all_data 和 data3

```python
pd.merge(all_data, data3, on='subject_id')
```

### Step 5. 按 'subject_id' 合并data1 和 data2（内连接） 

```python
pd.merge(data1, data2, on='subject_id', how='inner') 
```

### Step 6. 按 'subject_id' 合并data1 和 data2（外连接） 

```python
pd.merge(data1, data2, on='subject_id', how='outer') 
```

# Regiment 数据集

### Step 1. 根据raw_data创建数据集，并命名为regiment


```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

```python
regiment = pd.DataFrame(raw_data, columns=['regiment', 'company', 'name', 'preTestScore', 'postTestScore'])
regiment
```

### Step 2. Nighthawks 的平均 preTestScore?  

```python
regiment.loc[regiment['regiment'] == 'Nighthawks', 'preTestScore'].mean()
```

### Step 3. company列的统计数据

```python
regiment.groupby('company').describe()
```

### Step 4. 不同company的preTestScore均值?

```python
regiment.groupby('company').preTestScore.mean()
```

### Step 5. 不同regiment 不同 company的 preTestScores 均值

```python
regiment.groupby(['regiment', 'company']).preTestScore.mean()
```

### Step 6. 将上述多种索引转换为DataFrame

```python
regiment.groupby(['regiment', 'company']).preTestScore.mean().unstack()
```

### Step 7. 不同 regiment 不同 company的数量

```python
regiment.groupby(['regiment', 'company']).size()
```

### Step 8. 根据regiment分组，并打印 name 和 data

```python
for name, data in regiment.groupby(['regiment', 'company']):
    print(name)
    print(data)
```
